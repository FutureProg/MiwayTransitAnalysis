# Peel Open Data Streets
In this notebook we will be looking at the [Region of Peel Streets Open Dataset](http://opendata.peelregion.ca/data-categories/transportation/streets.aspx) with a focus primarily on the City of Mississauga. We will be exploring the features of the dataset and finding detailed labels of each. 

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

%matplotlib inline

pd.options.display.max_rows = 120

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df = gpd.read_file("/Users/Nick/Documents/Personal/Data/Peel/streets.geojson")
```

```python
df.columns
```

The original dataset contains information on cities outside of the region of peel: 

```python
str(df['L_MUN'].unique())
```

## Mississauga

```python
df_sauga = df[df['L_MUN'] == 'Mississauga']
```

```python
df_sauga['geometry'].plot(figsize=(20,20))
```

```python
df_sauga
```

ACC_PEDEST shows whether the road is pedestrian accessible or not. We can see that the highways are labeled as No Pedestrian Accessible as well as the airport roads and some private property in Clarkson. 

```python
df_sauga[df_sauga['ACC_PEDEST'] == 'N'].plot(figsize=(10,10))
```

Let's now take a look at all of the different street classes `ST_CLASS`:

```python
st_class_names = df_sauga['ST_CLASS'].unique()
st_class_names
```

```python
from matplotlib.lines import Line2D
fig, ax = plt.subplots(figsize=(15,15))
colors = ['orange', 'blue', 'green', 'black', 'purple', 'yellow', 'red']
# ax = df_sauga[df_sauga['ST_CLASS'] == st_class_names[0]].plot(figsize=(12,12), color=colors[0])
for index, cname in enumerate(st_class_names):
    df_sauga[df_sauga['ST_CLASS'] == cname].plot(ax=ax, figsize=(12,12), color=colors[index])
ax.set_axis_off()
lgnd = [Line2D([0], [0], color=colors[i], lw=3, label=cname) for i, cname in enumerate(st_class_names)]
ax.legend(lgnd, st_class_names, loc='upper left')
plt.show()
```
