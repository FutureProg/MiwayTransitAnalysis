```python
import pandas as pd
import folium
import overpy
api = overpy.Overpass()
```

```python
with open('/Volumes/Nicks1TB/Data/Mississauga/mississauga.osm') as f:
    osm_data = api.parse_xml(data=f.read(), encoding='utf-8', parser=None)
```

Given that we will be using OpenStreetMaps data, it can be assumed that this will not be an accurate 100% accurate assessment of the walkability between buildings and bus stops. 

```python
print(len([node.tags for node in osm_data.ways if 'addr:housenumber' in node.tags]))
```

```python
count = 0
good_ways = []
bad_ways = []
for x in osm_data.ways:
    try:
        x.nodes
        if 'aeroway' not in x.tags.keys() and 'motorway' not in x.tags['highway']\
        and x.tags['name'] == 'Winston Churchill Boulevard':
            good_ways.append(x)
    except:
        bad_ways.append(x)
print(len(bad_ways))
```

```python
len(good_ways)
```

```python
bad_highway_cnt = 0
bad_highways = []
for way in bad_ways:
    if 'highway' in way.tags.keys():
        bad_highway_cnt = bad_highway_cnt + 1
        if 'name' in way.tags.keys():
            bad_highways.append(way.tags['name'])
print(bad_highway_cnt)
bad_highways
```

```python
m = folium.Map(location=[good_ways[0].nodes[0].lat, good_ways[0].nodes[0].lon], zoom_start=11)
for way in good_ways[0:1000]:
    latlng = [(node.lat, node.lon) for node in way.nodes]
    folium.PolyLine( latlng, fill_color='#43d9de', weight=4, popup=way.tags ).add_to( m )
m
```
